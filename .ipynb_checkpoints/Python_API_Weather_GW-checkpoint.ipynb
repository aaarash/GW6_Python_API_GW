```python
import warnings
warnings.simplefilter('ignore')

import pandas as pd
import numpy as np
from scipy import stats

import matplotlib
import matplotlib.pyplot as plt

import requests
import time
import json

import seaborn as sns
sns.set_style('darkgrid')

%config InlineBackend.figure_format = 'retina'

# Turnning on plotting display in the notebook
%matplotlib inline

plt.style.use('fivethirtyeight')

from config_Weather_API import api_key

from citipy import citipy
```

```python
coordinates = []

cities = []

# Random Coordinates
lats = np.random.uniform(low=-70.000, high=80.000, size=1200)
lngs = np.random.uniform(low=-120.000, high=190.000, size=1800)
coordinates = zip(lats, lngs)

# nearest city 
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    
    if city not in cities:
        cities.append(city)

len(cities)
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
query_url = f"{url}appid={api_key}&units={units}&q="

country = []
lat = []
lng = []
humidity = []
cloud = []
wind = []
date = []
maxTemp = []

cityWeather_df = pd.DataFrame({"City": cities})                       
cityWeather_df["Country"] = ""
cityWeather_df["Lat"] = ""
cityWeather_df["Lng"] = ""
cityWeather_df["Humidity"] = "" 
cityWeather_df["Cloud"] = ""
cityWeather_df["Wind"] = ""
cityWeather_df["Date"] = ""
cityWeather_df["Max_Temp"] = ""

cityWeather_df.head()
```

```python
for index, row in cityWeather_df.iterrows():
    
    try:
    
        city = row["City"]
    
        response = requests.get(query_url + city).json()
    
        country = response['sys']['country']
        lat = response['coord']['lat']
        lng = response['coord']['lon']
        humidity = response['main']['humidity']
        cloud = response['clouds']['all']
        wind = response['wind']['speed']
        date = response['dt']
        maxTemp = response['main']['temp_max']
        

        cityWeather_df.at[index, 'Country'] = country
        cityWeather_df.at[index, 'Lat'] = lat
        cityWeather_df.at[index, 'Lng'] = lng
        cityWeather_df.at[index, 'Humidity'] = humidity
        cityWeather_df.at[index, 'Cloud'] = cloud
        cityWeather_df.at[index, 'Wind'] = wind
        cityWeather_df.at[index, 'Date'] = date
        cityWeather_df.at[index, 'Max_Temp'] = maxTemp
        print(f"Processing Record {index} | {city}\n {query_url}")
        
    except (KeyError, IndexError) as error:
        print(f"")
        print(f"") 
        pass 
 
cityWeather_df.dropna(how='any', inplace=True)
    
cityWeather_df.head()
```

```python
cityWeather_df["Lat"] = pd.to_numeric(cityWeather_df["Lat"])
cityWeather_df["Max_Temp"] = pd.to_numeric(cityWeather_df["Max_Temp"])

cityWeather_df.plot(kind = 'scatter', x="Lat", y="Max_Temp", alpha=1, zorder=3, figsize=(15, 7)) 
plt.title("Latitude vs. Temperature-Maximum (July 07, 2018)")
plt.xlabel("Latitude")
plt.ylabel("Temperature-Maximum in Farenheight")
plt.xlim(-80,80)
plt.show()
```
